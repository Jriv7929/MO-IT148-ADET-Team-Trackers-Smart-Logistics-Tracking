```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

num_records = 300  # Number of IoT data records

data = []

# Simple list of hubs and routes
warehouses = ["Manila Hub", "Cebu Hub", "Davao Hub"]
routes = ["Manila → Cebu", "Cebu → Davao", "Davao → Manila"]

for _ in range(num_records):
    speed = np.random.randint(0, 100)  # Vehicle speed
    fuel = np.random.randint(10, 100)  # Fuel level

    # Basic delivery status
    status = np.random.choice(["In Transit", "Idle", "Delivered"])

    # Simple delay logic (not too complex)
    delay = 0
    if speed < 20 or fuel < 20:
        status = "Delayed"
        delay = np.random.randint(5, 30)  # Delay in minutes

    record = {
        # Timestamp (last 24 hours)
        "timestamp": datetime.now() - timedelta(minutes=np.random.randint(0, 1440)),

        # Vehicle ID
        "vehicle_id": f"TRK{np.random.randint(100, 999)}",

        # Warehouse hub
        "warehouse_hub": np.random.choice(warehouses),

        # Route assignment
        "route": np.random.choice(routes),

        # GPS location (sample PH range)
        "latitude": round(np.random.uniform(12.0, 14.0), 6),
        "longitude": round(np.random.uniform(120.0, 122.0), 6),

        # Speed in km/h
        "speed_kmh": speed,

        # Fuel level (%)
        "fuel_level_percent": fuel,

        # Cargo temperature
        "cargo_temp_c": round(np.random.uniform(2.0, 10.0), 1),

        # Delay indicator
        "delay_minutes": delay,

        # Delivery status
        "delivery_status": status
    }

    data.append(record)

# Convert to DataFrame
df = pd.DataFrame(data)

# Save files
df.to_csv("logistics_data.csv", index=False)
df.to_json("logistics_data.json", orient="records")

# Show sample
print(df.head())
```

```python
# Call getTotalRecords() to see how many entries are currently saved
total_records = contract.functions.getTotalRecords().call()
print(f" Total records currently on the blockchain: {total_records}")
```

```python
# Send dummy transaction using contract's parameter fields
txn = contract.functions.storeLogisticsData(
    "SHIP999",          # _shipmentId
    "TRUCK-XYZ",        # _vehicleId
    "Jane Doe",         # _driverName
    "Express Route 4",  # _route
    "Central Hub 10",   # _warehouse
    "08:30 AM",         # _atd
    "01:15 PM",         # _ata
    "In Transit",       # _deliveryStatus
    "Pending",          # _podStatus
    15,                 # _predictedDelayMinutes (integer)
    "None",             # _emergencyAlert
    "Temp: 18C, Hum: 45%", # _environmentData
    "Fuel: 92%, Batt: OK", # _fleetData
    450                 # _inventoryCount (integer)
).transact({
    'from': web3.eth.default_account,
    'gas': 2000000      # Higher gas limit because the struct stores a lot of string data
})

# Wait for the local Ganache node to seal the block
tx_receipt = web3.eth.wait_for_transaction_receipt(txn)

print("✅ Custom logistic record successfully stored on the blockchain!")
print(f" Block Number: {tx_receipt['blockNumber']}")
#print(f" Transaction Hash: {tx_receipt['transactionHash'].hex()}")
```

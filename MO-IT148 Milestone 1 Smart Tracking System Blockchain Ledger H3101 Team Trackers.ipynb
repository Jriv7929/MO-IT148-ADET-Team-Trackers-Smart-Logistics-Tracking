```python
import time
import math

def clean_value(value):
    """
    Cleans CSV values before sending to blockchain
    """

    # Handle NaN values
    if pd.isna(value):
        return ""

    # Convert floats cleanly
    if isinstance(value, float):
        return format(value, '.6f')

    return str(value)


def send_logistics_data(row):

    try:

        txn = contract.functions.storeLogisticsData(

            clean_value(row["Shipment_ID"]),
            clean_value(row["Vehicle_ID"]),
            clean_value(row["Route"]),
            clean_value(row["Timestamp"]),
            clean_value(row["GPS_Latitude"]),
            clean_value(row["GPS_Longitude"]),
            clean_value(row["Vehicle_Speed_kmh"]),
            clean_value(row["RFID_Tag_ID"]),
            clean_value(row["RFID_Scan_Status"]),
            clean_value(row["Package_Status"]),
            clean_value(row["Temperature_C"]),
            clean_value(row["Alert_Status"])

        ).transact({
            "from": web3.eth.default_account,
            "gas": 3000000
        })

        receipt = web3.eth.wait_for_transaction_receipt(txn)

        print(f"✅ Stored Shipment: {row['Shipment_ID']}")
        print(f"Transaction Hash: {receipt.transactionHash.hex()}")

    except Exception as e:

        print(f"❌ Error storing shipment {row['Shipment_ID']}")
        print(e)
```

```python
# Send ONLY the first 2 rows from the CSV file

for index, row in df.head(2).iterrows():
    send_logistics_data(row)
    time.sleep(1)  # Delay between transactions

print("✅ 2 logistics records successfully stored on the blockchain!")
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# ==========================================
# SMART LOGISTICS TRACKING DATA GENERATOR
# ==========================================

# Number of rows
num_records = 200

# ------------------------------------------
# Sample Data
# ------------------------------------------

vehicle_ids = [f"TRK-{100+i}" for i in range(15)]

routes = [
    "Manila to Quezon City",
    "Pasig to Makati",
    "Taguig to Cavite",
    "Bulacan to Manila",
    "Laguna to Pasay",
    "Cebu City to Mandaue",
    "Davao to Tagum"
]

delivery_statuses = [
    "Delivered",
    "In Transit",
    "Delayed",
    "Cancelled"
]

data = []

# ==========================================
# GENERATE DATA
# ==========================================

for i in range(num_records):

    # Random timestamp
    timestamp = datetime.now() - timedelta(
        days=np.random.randint(0, 30),
        hours=np.random.randint(0, 24),
        minutes=np.random.randint(0, 60)
    )

    # GPS SENSOR DATA
    gps_latitude = round(np.random.uniform(14.40, 14.85), 6)
    gps_longitude = round(np.random.uniform(120.90, 121.20), 6)
    vehicle_speed = round(np.random.uniform(20, 100), 2)

    # RFID TAG DATA
    rfid_tag = f"RFID-{5000+i}"
    rfid_scan_status = random.choice([
        "Scanned",
        "Not Scanned"
    ])

    package_status = random.choice([
        "Loaded",
        "In Transit",
        "Delivered",
        "Missing"
    ])

    # TEMPERATURE SENSOR DATA
    temperature = round(np.random.uniform(2, 35), 2)

    # DELIVERY STATUS
    delivery_status = random.choice(delivery_statuses)

    # ALERT STATUS LOGIC
    if temperature > 30:
        alert_status = "Temperature Alert"

    elif rfid_scan_status == "Not Scanned":
        alert_status = "RFID Scan Missing"

    elif vehicle_speed < 30:
        alert_status = "Possible Delay"

    else:
        alert_status = "Normal"

    # ======================================
    # RECORD
    # ======================================

    record = {

        # Basic Shipment Information
        "Shipment_ID": f"SHP-{1000+i}",
        "Vehicle_ID": random.choice(vehicle_ids),
        "Route": random.choice(routes),
        "Timestamp": timestamp.strftime("%Y-%m-%d %H:%M:%S"),

        # GPS Sensors Data
        "GPS_Latitude": gps_latitude,
        "GPS_Longitude": gps_longitude,
        "Vehicle_Speed_kmh": vehicle_speed,

        # RFID Tags Data
        "RFID_Tag_ID": rfid_tag,
        "RFID_Scan_Status": rfid_scan_status,
        "Package_Status": package_status,

        # Temperature Sensor Data
        "Temperature_C": temperature,

        # Monitoring Status
        "Alert_Status": alert_status
    }

    data.append(record)

# ==========================================
# CREATE DATAFRAME
# ==========================================

df = pd.DataFrame(data)

# ==========================================
# SAVE CSV FILE
# ==========================================

df.to_csv("smart_logistics_tracking.csv", index=False)

# ==========================================
# OUTPUT
# ==========================================

print("CSV file generated successfully!")
print("Rows:", len(df))
print("Columns:", len(df.columns))

print("\nFirst 5 Rows:")
print(df.head())
```

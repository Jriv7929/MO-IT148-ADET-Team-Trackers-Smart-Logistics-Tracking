```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random

# Number of records
num_records = 300

# Sample data
vehicle_ids = [f"TRK-{100+i}" for i in range(20)]
drivers = [
    "John Cruz", "Maria Santos", "Kevin Reyes", "Anne Lopez",
    "Mark Dela Cruz", "Paul Garcia", "Liza Mendoza"
]

routes = [
    "Manila to Quezon City",
    "Pasig to Makati",
    "Taguig to Cavite",
    "Bulacan to Manila",
    "Laguna to Pasay",
    "Cebu City to Mandaue",
    "Davao to Tagum"
]

warehouses = [
    "Warehouse A - Manila",
    "Warehouse B - Pasig",
    "Warehouse C - Laguna",
    "Warehouse D - Cebu"
]

delivery_statuses = [
    "Delivered",
    "In Transit",
    "Delayed",
    "Cancelled"
]

emergency_alerts = [
    "None",
    "Flat Tire",
    "Engine Failure",
    "Accident",
    "Heavy Traffic",
    "Road Closure"
]

pod_status = ["Received", "Pending", "Failed"]

data = []

for i in range(num_records):

    # Generate random departure time
    atd = datetime.now() - timedelta(
        days=np.random.randint(0, 30),
        hours=np.random.randint(0, 24),
        minutes=np.random.randint(0, 60)
    )

    # Simulated travel duration
    travel_hours = np.random.randint(1, 12)

    # Delay prediction
    predicted_delay = random.choice([0, 5, 10, 15, 20, 30, 45])

    # Actual arrival time
    ata = atd + timedelta(hours=travel_hours, minutes=predicted_delay)

    # Environment monitoring
    temperature = round(np.random.uniform(18, 40), 2)
    humidity = round(np.random.uniform(40, 95), 2)

    # Speed and fuel
    speed = round(np.random.uniform(30, 100), 2)
    fuel_level = round(np.random.uniform(10, 100), 2)

    # Inventory and warehouse
    inventory_count = np.random.randint(10, 500)

    # Delivery status logic
    status = random.choice(delivery_statuses)

    # Emergency alert logic
    emergency = random.choices(
        emergency_alerts,
        weights=[70, 5, 5, 3, 12, 5],
        k=1
    )[0]

    # Proof of delivery
    pod = random.choice(pod_status)

    record = {
        "Shipment_ID": f"SHP-{1000+i}",
        "Vehicle_ID": random.choice(vehicle_ids),
        "Driver_Name": random.choice(drivers),
        "Route": random.choice(routes),
        "Warehouse": random.choice(warehouses),

        # Time Monitoring
        "ATD": atd.strftime("%Y-%m-%d %H:%M:%S"),
        "ATA": ata.strftime("%Y-%m-%d %H:%M:%S"),

        # Delivery
        "Delivery_Status": status,
        "PoD_Status": pod,

        # Delay Prediction
        "Predicted_Delay_Minutes": predicted_delay,

        # Emergency Alerts
        "Emergency_Alert": emergency,

        # Environment Monitoring
        "Temperature_C": temperature,
        "Humidity_%": humidity,

        # Fleet Management
        "Vehicle_Speed_kmh": speed,
        "Fuel_Level_%": fuel_level,

        # Inventory Management
        "Inventory_Count": inventory_count
    }

    data.append(record)

# Convert to DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("logistics_data.csv", index=False)

print("CSV file generated successfully!")
print(df.head())
```

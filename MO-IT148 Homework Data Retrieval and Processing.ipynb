```python
# ==============================
# WEEK 6: DATA RETRIEVAL AND PROCESSING
# ==============================
```

```python
# 1. Get total number of stored records
total_records = contract.functions.getTotalRecords().call()
print("Total Stored Records:", total_records)
```

```python
# 2. Fetch all stored logistics records
retrieved_data = []

for i in range(total_records):
    record = contract.functions.getRecord(i).call()

    retrieved_data.append({
        "Shipment_ID": record[0],
        "Vehicle_ID": record[1],
        "Route": record[2],
        "Timestamp": record[3],
        "GPS_Latitude": record[4],
        "GPS_Longitude": record[5],
        "Vehicle_Speed_Kmh": record[6],
        "RFID_Tag_ID": record[7],
        "RFID_Scan_Status": record[8],
        "Package_Status": record[9],
        "Temperature_C": record[10],
        "Alert_Status": record[11]
    })
```

```python
# 3. Convert retrieved data to DataFrame
retrieved_df = pd.DataFrame(retrieved_data)

print("\n✅ Retrieved Blockchain Data:")
print(retrieved_df.head())
```

```python
# 4. Function to extract numeric values
def extract_numeric(value):
    if pd.isna(value):
        return None

    value = str(value)
    number = re.findall(r"[-+]?\d*\.\d+|\d+", value)

    if number:
        return float(number[0])
    return None
```

```python
# 5. Convert numerical columns
numeric_columns = [
    "GPS_Latitude",
    "GPS_Longitude",
    "Vehicle_Speed_Kmh",
    "Temperature_C"
]

for col in numeric_columns:
    retrieved_df[col] = retrieved_df[col].apply(extract_numeric)
```

```python
# 6. Identify missing values
print("\nMissing Values Before Cleaning:")
print(retrieved_df.isnull().sum())
```
